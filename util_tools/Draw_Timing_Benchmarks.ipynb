```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_yolov3 = 'fastly_benchmark_yolov3.txt'
file_yolov4 = 'fastly_benchmark_yolov4.txt'

timer_stamps_yolov3 = []
timer_stamps_yolov4 = []

with open(file_yolov3, 'r') as file:
    lines = file.readlines()
    for line in lines:
        timer_stamps_yolov3.append(float(line))
        
with open(file_yolov4, 'r') as file:
    lines = file.readlines()
    for line in lines:
        timer_stamps_yolov4.append(float(line))
```

```python
# define data values
xaxis = [i for i in range(20)]
plt.figure(figsize=(30,20))
plt.plot(xaxis, timer_stamps_yolov3, label = "YOLOv3", linewidth = 5)  # Plot the chart
plt.plot(xaxis, timer_stamps_yolov4, label = "YOLOv4", linewidth = 5)  # Plot the chart
plt.xticks(range(20))
plt.xlabel('Number of Image', fontsize=42)
plt.ylabel('Time (s)', fontsize=42)
plt.title('Runtime Statistics of YOLOv3 and YOLOv4',fontsize=42, pad=35)
plt.rc('xtick',labelsize=42)
plt.rc('ytick',labelsize=42)
plt.legend(loc="lower right", fontsize = 42)
# plt.show()  # display
plt.savefig('timing_benchmark.png',dpi=300)
```
