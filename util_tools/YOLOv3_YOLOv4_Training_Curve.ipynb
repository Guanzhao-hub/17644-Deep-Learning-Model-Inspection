```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file1 = 'train_log_WARMUP2_BATCH16_TRAIN_LR_INIT0.001_TRAIN_LR_END1e-06_TRAIN_EPOCHS200.txt'

total_val_yaxis_yolov3 = []
substring = "total_val_loss:"

with open(file1, 'r') as file:
    lines = file.readlines()
    for line in lines:
        if substring in line:
            total_val_loss = float(line.split(substring)[-1].strip())
            total_val_yaxis_yolov3.append(total_val_loss)
            
total_val_yaxis_yolov3 = total_val_yaxis_yolov3[2:]
print(len(total_val_yaxis_yolov3))
```

```python
file_name = 'nohup.txt'
total_val_yaxis_yolov4 = []
substring = "total_val_loss:"

with open(file_name, 'r') as file:
    lines = file.readlines()
    for line in lines:
        if substring in line:
            total_val_loss = float(line.split(substring)[-1].strip())
            total_val_yaxis_yolov4.append(total_val_loss)
total_val_yaxis_yolov4 = total_val_yaxis_yolov4[2:]
print(len(total_val_yaxis_yolov4))
```

```python
# define data values
xaxis = [i for i in range(198)]
plt.figure(figsize=(30,20))
plt.plot(xaxis, total_val_yaxis_yolov3, label = "YOLOv3", linewidth = 5)  # Plot the chart
plt.plot(xaxis, total_val_yaxis_yolov4, label = "YOLOv4", linewidth = 5)  # Plot the chart
plt.xlabel('Epoch', fontsize=42)
plt.ylabel('Loss', fontsize=42)
plt.title('Training Curve of YOLOv3 and YOLOv',fontsize=42, pad=35)
plt.rc('xtick',labelsize=42)
plt.rc('ytick',labelsize=42)
plt.legend(loc="upper right", fontsize = 42)
# plt.show()  # display
plt.savefig('Training_Curve_YOLOv3_YOLOv4',dpi=300)
```
