```python
import matplotlib.pyplot as plt
import numpy as np


```

```python
file1 = open('EPOCHS400.txt', 'r')

total_val_yaxis = []
conf_val_yaxis=[]
substring = "total_val_loss"
while True: 
    line = file1.readline()
    if substring in line:
        #index=line.find(substring,0)
        #line[80:90]
        total_val_loss = float(line.split(" ")[-1].strip())
        conf_val_loss= line.split(" ")[7].strip()
        conf_val_loss = float(conf_val_loss[:-1])
        total_val_yaxis.append(total_val_loss)
        conf_val_yaxis.append(conf_val_loss)
```

```python
# define data values
xaxis = [i for i in range(400)]
plt.figure(figsize=(30,20),dpi=300)
plt.plot(xaxis, total_val_yaxis, label = "line 1")  # Plot the chart
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()  # display
```
