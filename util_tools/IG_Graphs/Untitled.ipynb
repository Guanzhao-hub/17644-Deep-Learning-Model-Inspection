```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file1 = 'training_log_early_stopping.txt'
training_loss = []

substring = "- loss:"

with open(file1, 'r') as file:
    lines = file.readlines()
    for line in lines:
        if substring in line:
            total_val_loss = float(line.split(substring)[-1].strip().split(' ')[0])
            training_loss.append(total_val_loss)
print(len(training_loss))
print(training_loss)
```

```python
accuracy = []

substring = "- accuracy: "

with open(file1, 'r') as file:
    lines = file.readlines()
    for line in lines:
        if substring in line:
            total_val_loss = float(line.split(substring)[-1].strip().split(' ')[0])
            accuracy.append(total_val_loss)
print(len(accuracy))
print(accuracy)
```

```python
val_loss = []

substring = "- val_loss: "

with open(file1, 'r') as file:
    lines = file.readlines()
    for line in lines:
        if substring in line:
            total_val_loss = float(line.split(substring)[-1].strip().split(' ')[0])
            val_loss.append(total_val_loss)
print(len(val_loss))
print(val_loss)
```

```python
val_accuracy = []

substring = "- val_accuracy: "

with open(file1, 'r') as file:
    lines = file.readlines()
    for line in lines:
        if substring in line:
            total_val_loss = float(line.split(substring)[-1].strip().split(' ')[0])
            val_accuracy.append(total_val_loss)
print(len(val_accuracy))
print(val_accuracy)
```

```python
# define data values
xaxis = [i for i in range(14)]
plt.figure(figsize=(30,20))
plt.plot(xaxis, val_accuracy, label = "Validation Accuracy", linewidth = 5)  # Plot the chart
plt.plot(xaxis, accuracy, label = "Training Accuracy", linewidth = 5)  # Plot the chart
plt.xlabel('Epoch', fontsize=42)
plt.ylabel('Accuracy', fontsize=42)
plt.title('Accuracy vs. Epoch',fontsize=42, pad=35)
plt.rc('xtick',labelsize=42)
plt.rc('ytick',labelsize=42)
plt.legend(loc="upper right", fontsize = 42)
# plt.show()  # display
plt.savefig('Traning Accuracy vs EPOCH Regularization L1',dpi=300)
```

```python
# define data values
xaxis = [i for i in range(14)]
plt.figure(figsize=(30,20))
plt.plot(xaxis, training_loss, label = "Train Loss", linewidth = 5)  # Plot the chart
plt.plot(xaxis, val_loss, label = "Val Loss", linewidth = 5)  # Plot the chart
plt.xlabel('Epoch', fontsize=42)
plt.ylabel('Loss', fontsize=42)
plt.title('Training Curve of YOLOv3 and YOLOv',fontsize=42, pad=35)
plt.rc('xtick',labelsize=42)
plt.rc('ytick',labelsize=42)
plt.legend(loc="upper right", fontsize = 42)
# plt.show()  # display
plt.savefig('Traning Curve of Mask Classification Regularization L1',dpi=300)
```
